## b. Как изменился средний чек транзакции после введения коронавирусных ограничений? Как он изменился для пользователей, покупающих с промокодом и без? (параметр «promo_activated»)

#### Почистим данные, чтобы избежать деление на ноль (возьмем данные только с transactions > 0).

```python
df_new = df[df['transactions'] > 0].sort_values('source')
df_new 
```

```python
df_new.shape
```

#### Добавим признак средний чек(revenue/transactions)

```python

average_check = []
for i, row in df_new.iterrows():
    average_check.append(row['revenue']/row['transactions'])
df_new['average_check'] = average_check
df_new.head()
```

#### Дата введения коронавирусных ограничений 16 марта 2020 г.

Проверка данных на нормальность:

```python
from scipy.stats import shapiro
alpha = 0.05
st = shapiro(df_new.average_check)
st
```

```python
print('Distribution is {}normal\n'.format( {True:'not ', False:''}[st[1] < alpha]));
```

Распределение не является нормальным, поэтому для оценки статистической значимости различий будем использовать критерий Манн-Уитни.

```python
df_before = df_new[df_new['date'] < '2020-03-16'].sort_values('date')
df_before 
```

```python
df_after = df_new[df_new['date'] >= '2020-03-16'].sort_values('date')
df_after 
```

```python
import scipy.stats as stats
X1= np.array(df_before.average_check)
Y1= np.array (df_after.average_check)
stats.mannwhitneyu(X1, Y1)
```

Принимаем alpha = 0.05. pvalue < alpha, имеются статистически значимые различия между выборками до коронавируса и после.

```python
df_before['average_check'].mean()
```

```python
df_before[df_before['promo_activated'] == 'no']['average_check'].mean()
```

```python
df_before[df_before['promo_activated'] == 'yes']['average_check'].mean()
```

```python
df_after['average_check'].mean()
```

```python
df_after[df_after['promo_activated'] == 'no']['average_check'].mean()
```

```python
df_after[df_after['promo_activated'] == 'yes']['average_check'].mean()
```

```python
X1_no= np.array(df_before[df_before['promo_activated'] == 'no']['average_check'])
Y1_no= np.array (df_after[df_after['promo_activated'] == 'no']['average_check'])
stats.mannwhitneyu(X1_no, Y1_no)
```

```python
X1_yes= np.array(df_before[df_before['promo_activated'] == 'yes']['average_check'])
Y1_yes= np.array (df_after[df_after['promo_activated'] == 'yes']['average_check'])
stats.mannwhitneyu(X1_yes, Y1_yes)
```

### Вывод:средний чек транзакции после введения коронавирусных ограничений снизился на (1-1193.63/1292.79)*100% = 7,67 %; для пользователей, покупающих с промокодом снизился на (1-1206.78/1276.38)*100% = 5,45 %; без промокода снизился на (1-1179.51/1309.13)*100% = 9,90 %

## c. Можно ли с уверенностью в 95% сказать, что CR (коэффициент конверсии в
## транзакцию) в выходные дни отличается от CR в будние дни?

Добавим признак CR_transactions(коэффициент конверсии в транзакцию). CR_transactions = transactions/visits*100%

```python
CR_transactions = []
for i, row in df.iterrows():
    CR_transactions.append(round(row['transactions']/row['visits']*100, 2))
df['CR_transactions'] = CR_transactions
df
```

```python
import calendar
calendar.prcal(2020)
```

Добавим признак weekday

```python
df['weekday'] = pd.to_datetime(df['date']).apply(lambda x: x.weekday())
df.sort_values('date')
```

Считаем, что выходные дни: суббота и воскресенье( 'weekday' = 5 или 6).

```python
df_weekends = df[df['weekday'].isin([5, 6])].sort_values('date')
df_weekends 
```

```python
df_weekdays = df[df['weekday'].isin([0, 1, 2, 3, 4])]
df_weekdays
```

#### Проверка на нормальность распределения

```python
from scipy.stats import shapiro
alpha = 0.05
st = shapiro(df.CR_transactions)
st
```

```python
print('Distribution is {}normal\n'.format( {True:'not ', False:''}[st[1] < alpha]));
```

У нас не соблюдается условие нормальности делаем вывод, что для проверки отличается ли коэффициент конверсии в выходные и будние дни будем использовать критерий Манна-Уитни.

 Непараметрический критерий  Манн-Уитни

```python
X2= np.array(df_weekends.CR_transactions)
Y2= np.array (df_weekdays.CR_transactions)
stats.mannwhitneyu(X2, Y2)
```

alpha = 0.05. pvalue < alpha, имеются статистически значимые различия между выборками в выходные и будни.

### Ответ: C уверенностью в 95%  мы можем сказать, что CR (коэффициент конверсии в
### транзакцию) в выходные дни статистически значимо отличается от CR в будние дни
